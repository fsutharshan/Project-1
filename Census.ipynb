```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_path1 = "muniestbycounty_byyr.csv"
population_df=pd.read_csv(csv_path1,encoding="utf-8",sep=';')
population_df.head()
```

```python
csv_path2 = "uszips.csv"
zip_df = pd.read_csv(csv_path2,encoding="utf-8")
zip_df.head()
```

```python
new_zip=zip_df[["zip","lat","lng","city","state_name","county_name"]]
new_zip.head()
```

```python
nc_df = new_zip.loc[new_zip["state_name"] == "North Carolina"]
nc_df= nc_df.rename(columns={"county_name":"County"})
nc_df.head()
```

```python
census_df=pd.merge(nc_df,population_df,on='County')
census_df.head()
```
