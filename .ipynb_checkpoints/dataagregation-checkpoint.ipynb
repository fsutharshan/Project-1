```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
from scipy import stats
```

```python
data1 = pd.read_csv('hospitals/output_data/county_metrics')
data2 = pd.read_csv('hospitals/output_data/UShospitals.csv')
data3 = pd.read_csv('hospitals/output_data/political.csv')
data4 = pd.read_csv('Resources/NC_Population_county.csv')
covid_data = pd.DataFrame(data1)
hospitals =pd.DataFrame(data2)
politics = pd.DataFrame(data3)
population = pd.DataFrame(data4)
```

```python
covid_data=covid_data.rename(columns={'ID':'fips'})
del covid_data['Unnamed: 0']
covid_data.head()
```

```python
del hospitals['Unnamed: 0']
del hospitals['County']
del hospitals['State']
hospitals.head()
```

```python
del politics['FIPS.1']
politics =politics.rename(columns={'FIPS':'fips'})
politics.head()
```

```python
df=pd.merge(covid_data,hospitals,on ='fips',how='left')
df = pd.merge(df,politics,on='fips',how='left')
```

```python
df2 = df.loc[(df['R Squared']  >0.2)&(df['County']!='Unknown')]
              &(df['State']=='North Carolina')]
df2['illness']=df2['Average Rating']*df2['Hospitals']
df2=df2.sort_values('Tau', ascending =False)
df2.head(40)
```

```python
bins=[]
step =0.5
total = int(35/step)
for x in range (total):
    bins.append(x*step)
plt.hist(df2['Tau'],bins=bins)
plt.show()
```

```python
plt.scatter(df2['Average Rating'],df2['Tau'])
# plt.ylim(4,10)
hold_df = df2.dropna(subset=('Average Rating','Tau'))
corr=stats.pearsonr(hold_df['Average Rating'],hold_df['Tau'])
print(corr[0])
(slope,intercept,x, y, z)=stats.linregress(hold_df['Average Rating'],hold_df['Tau'])
xval = np.arange(0,5,0.2)
yval=slope*xval+intercept
plt.plot(xval,yval)
plt.show()
```

```python
plt.scatter(df2['County Beds'],df2['Tau'])
# plt.ylim(4,10)
# plt.xlim(0,500)
hold_df = df2.dropna(subset=('County Beds','Tau'))
corr=stats.pearsonr(hold_df['County Beds'],hold_df['Tau'])
print(corr[0])
(slope,intercept,x, y, z)=stats.linregress(hold_df['County Beds'],hold_df['Tau'])
print(slope)
xval = np.arange(0,max(df2['County Beds']),5)
yval=slope*xval+intercept
plt.plot(xval,yval)
plt.show()
```

```python
plt.scatter(df2['Hospitals'],df2['Tau'])
# plt.ylim(4,10)
# plt.xlim(0,10)
hold_df = df2.dropna(subset=('Hospitals','Tau'))
corr=stats.pearsonr(hold_df['Hospitals'],hold_df['Tau'])
print(corr[0])
plt.show()
```

```python
plt.scatter(df2['illness'],df2['Tau'])
# plt.ylim(4,10)
hold_df = df2.dropna(subset=('illness','Tau'))
corr=stats.pearsonr(hold_df['illness'],hold_df['Tau'])
print(corr[0])
plt.show()
```

```python
plt.scatter(df2['R/D Spectrum'],df2['Tau'])
# plt.ylim(4,10)
hold_df = df2.dropna(subset=('R/D Spectrum','Tau'))
corr=stats.pearsonr(hold_df['R/D Spectrum'],hold_df['Tau'])
print(corr[0])
plt.show()
```
