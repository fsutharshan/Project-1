```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('countypres_2000-2016.csv')
df = pd.DataFrame(data)
df.head()
```

```python
df['percent']=df['candidatevotes']/df['totalvotes']
df.head()
```

```python
fips = df['FIPS'].value_counts()
years = df['year'].value_counts()
```

```python
results={'State':[],'County':[],'FIPS':[],'Year':[],'R/D Spectrum':[]}
for fps in fips.index:
    for yr in years.index:
        hold_df = df.loc[(df['FIPS']==fps)&(df['year']==yr)]
        rep = hold_df.loc[hold_df['party']=='republican']
        rep = rep.reset_index()
        dem= hold_df.loc[hold_df['party']=='democrat']
        dem = dem.reset_index()
        try:
            results['State'].append(dem['state'][0])
            results['County'].append(dem['county'][0])
            results['R/D Spectrum'].append(rep['percent'][0]-dem['percent'][0])
        except:
            results['State'].append(0)
            results['County'].append(0)
            results['R/D Spectrum'].append(0)
        results['FIPS'].append(fps)
        results['Year'].append(yr)
        
```

```python
df2 = pd.DataFrame(results)
df2.head()
```

```python
# hdf=df.loc[df['FIPS']==8014]
hdf = df2.sort_values(by=['R/D Spectrum'])
hdf = df2.groupby('FIPS').mean()
hdf['FIPS']= hdf.index
hdf.to_csv('../output_data/political.csv')
hdf.head(30)
```
